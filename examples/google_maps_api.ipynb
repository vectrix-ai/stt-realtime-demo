# Route Duration Calculation
 In this notebook, we'll demonstrate the Google distance matrix API to calculate distances between destinations. We can choose transportation type and use a large language model to calculate distances.

 
### API Documentation

The Google Distance Matrix API accepts the following parameters:

- **origins**: A single location or a list of locations (string, dict, list, or tuple)
  - Can be addresses, Place IDs (prepended with 'place_id:'), or latitude/longitude values
  - Used as starting points for distance and time calculations

- **destinations**: A single location or a list of locations (string, dict, list, or tuple)
  - Can be addresses, Place IDs (prepended with 'place_id:'), or latitude/longitude values
  - Used as endpoints for distance and time calculations

- **mode**: (string) Specifies the transport mode
  - Valid values: "driving", "walking", "transit", "bicycling"

- **language**: (string) The language for results

- **avoid**: (string) Features to avoid in route calculation
  - Valid values: "tolls", "highways", "ferries"

- **units**: (string) Unit system for results
  - Valid values: "metric", "imperial"

- **departure_time**: (int or datetime.datetime) Desired departure time

- **arrival_time**: (int or datetime.datetime) Desired arrival time for transit directions
  - Note: Cannot specify both departure_time and arrival_time

- **transit_mode**: (string or list of strings) Preferred transit modes (only for transit requests)
  - Valid values: "bus", "subway", "train", "tram", "rail" ("rail" is equivalent to ["train", "tram", "subway"])

- **transit_routing_preference**: (string) Preferences for transit requests
  - Valid values: "less_walking", "fewer_transfers"

- **traffic_model**: (string) Predictive travel time model (only for driving mode with departure_time)
  - Valid values: "best_guess", "optimistic", "pessimistic"

- **region**: (string) Preferred region for geocoder (ccTLD code)

**Return type**: Matrix of distances, with results in rows (each row contains one origin paired with each destination)




```python
import googlemaps
from datetime import datetime
import dotenv
import os

dotenv.load_dotenv()

gmaps = googlemaps.Client(key=os.getenv("GOOGLE_MAPS_API_KEY"))

gmaps.distance_matrix(
    origins=["New York City, NY"],
    destinations=["Chicago, IL"],
    mode="driving",
    language="en",
    avoid="tolls",
    units="metric",
    departure_time=datetime.now(),
)
```
